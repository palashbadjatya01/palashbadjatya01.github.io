```python
# 2) Total square footage as a function of building status ("Bldg Status")
```

```python
# As tought in lecture I started with converting the data into group using the groupby function keeping Bldg
#Status and I used the sum function to add up the total square footage alnog the individual buildng status status
# i.e "In Use, In Progress or Abandon"

TSQF = buildings2.groupby("Bldg Status")["Square Footage"].sum()
TSQF.plot(x="Building Status", y="Total Square Footage", kind='bar')
plt.title('Total Square footage as a function of Buildng Status')
```

```python
# 3) Median square footage in each Year Constructed as a function of Year Acquired
```

```python
# As tought in lecture I started with converting the data into group using the groupby function but used only two parameters
# in it those were Year Constructed and Square Footage Mean but later on realised to also include Year Acquired because 
# question asked to plot keeping Year acquired on the x axis so after looking at the lectures and searching some examples
# on the internet I found this Workaround to use groupby one time and create a combined table having all the three required
# values in it. Initially I was curious to know that how to read Kernel Density Estimation Plot and Hexbin Plot because those 
# were new to me but later on I realised that might not be a good ft for data like this so settled with scatter plot.

#For calculating median and I used the median function on square footage 

MSQF = buildings2.groupby("Year Constructed")[["Year Acquired","Square Footage"]].median()
MSQF.plot(x="Year Acquired",y="Square Footage",kind='scatter',figsize=(20,10))
plt.title('Median square footage in each Year Constructed as a function of Year Acquired')
```
